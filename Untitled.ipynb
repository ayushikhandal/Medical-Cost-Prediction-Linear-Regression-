```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'insurance.csv')
data.head()
```

```python
#let's see some information regarding the dataset
data.info()
```

```python
#statistical overview
data.describe()
```

```python
#checking for duplicates
data.loc[data.duplicated()]
```

```python
#let's drop 
data=data.drop(581)
```

```python
data.loc[data.duplicated()]
```

```python
#let's see all columns
data.columns
```

```python
#converting categorical to int 
data['sex'] = data['sex'].astype('category')
data['sex'] = data['sex'].cat.codes
```

```python
data.head()
```

```python
data['smoker'] = data['smoker'].astype('category')
data['smoker'] = data['smoker'].cat.codes

data['region'] = data['region'].astype('category')
data['region'] = data['region'].cat.codes
```

```python
data.head()
```

```python
#Checking if any null values exist
data.isnull().sum()
```

```python
#creating the Model
y=data['charges']
```

```python
X = data.drop(columns = 'charges')
```

```python
X
```

```python
y
```

```python
#splititng the dataset
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
b = model.intercept_
b
```

```python
#prediticting on train model
y_train_predict = model.predict(X_train)
```

```python
y_train_predict
```

```python
#Comparing y_train vs predicted y_train
plt.scatter(y_train,y_train_predict)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.show()
```

```python
#R^2 score means that our model how well predict
from sklearn.metrics import r2_score
r2_score(y_train,y_train_predict)
```

```python
#This time trying with testing data
y_test_predict = model.predict(X_test)
```

```python
y_test_predict
```

```python
#Comparing y_test vs predicted y_test
plt.scatter(y_test,y_test_predict)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.show()
```

```python
#As you see, our train set and test set accuracy is so similar. 
#Also when you look two graph, you can see again
r2_score(y_test,y_test_predict)
```
